# Primeiros passos em machine learning

Dataset Iris: https://archive.ics.uci.edu/ml/machine-learning-databases/iris/

Como importar dados de um dataset, analisar de forma básica e criar um modelo de árvores de decisão.

Começamos importando os dados com o pandas. Para facilitar importei direto do repositório da UCI. O dataset Iris é composto de dados a respeito de flores Iris. As features são: comprimento da sépala, largura da sépala, comprimento da pétala e largura da pétala. O objetivo é classificar com base nessas features entre três espécies de íris: Iris setosa, Iris virginica e Iris versicolor.

```python
import pandas as pd # Importa o Pandas

arquivo = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data' # Nome e local do arquivo
nomes_col = ['comp_sepala', 'larg_sepala', 'comp_petala', 'larg_petala', 'especie'] # Títulos das colunas

df = pd.read_csv(arquivo, names=nomes_col) # Cria um dataframe do pandas com o arquivo e nomes de colunas especificados
```

```python
import seaborn as sns # Importa a biblioteca Seaborn (para plot)
sns.set(style='ticks') # Define o "estilo" que será usado nos gráficos
```

```python
y = df['especie'] # Cria o dataframe "objetivo", que é o com as espécies que queremos prever
X = df.drop(['especie'], axis=1) # Cria um dataframe com os dados de entrada para as previsões
```

```python
from sklearn.model_selection import train_test_split # Importa parte do sklearn para dividir os dados de entrada entre treino e teste
```

```python
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, random_state=0) # Divide os dados de entrada entre treino e teste
```

```python
from sklearn.tree import DecisionTreeClassifier # Importa o modelo de classificação de Árvores de Decisão
```

```python
modelo = DecisionTreeClassifier() # Atribui o modelo a uma variável
```

```python
modelo.fit(X_treino, y_treino) # Treina o nosso modelo
```

```python
modelo.score(X_teste, y_teste) # Verifica a acurácia do modelo
```

```python
modelo.predict(X_teste) # Gera previsões com os dados de entrada de teste
```
